# Bài tập phân loại lĩnh vực khoa học

Đôi khi rất khó để theo kịp tốc độ tiến bộ của khoa học, với những lĩnh vực khoa học mới (tương đối) có tên gọi nghe rất thú vị bao gồm proteomics, sinh học lượng tử và thần kinh ngôn ngữ học tính toán. Nhiệm vụ của bạn trong bài tập này là viết một "bộ phân loại" khoa học đơn giản, nhận tên của một lĩnh vực khoa học làm đầu vào và in ra một nhận xét (súc tích) như sau:

* Nếu tên **kết thúc** bằng *omics*, mã của bạn nên in `Life science hippy!`
* Nếu tên **bắt đầu** bằng *comp* hoặc *info*, mã của bạn nên in `Computing ftw!`
* Nếu tên **kết thúc** bằng *y*, mã của bạn nên in `Au naturel!`
* Nếu không thỏa mãn điều kiện nào ở trên, mã của bạn nên in `Too new to keep up!`

Lưu ý rằng đối với đầu vào khớp với nhiều điều kiện này, điều kiện khớp được liệt kê đầu tiên sẽ được áp dụng (và chỉ in ra một thông báo). Bạn có thể giả định rằng đầu vào sẽ hoàn toàn bao gồm các ký tự viết thường và dấu cách.

Chương trình của bạn nên hoạt động như sau:

## **Gợi ý**

Sau này bạn sẽ học cách trích xuất tiền tố và hậu tố từ chuỗi, nhưng đối với mục đích của chúng ta ở đây, cách an toàn nhất để đảm bảo rằng một kết quả khớp chuỗi con ở đầu hoặc cuối chuỗi là chèn trước một ký tự đặc biệt được liên kết duy nhất với vị trí đó trong chuỗi (ví dụ: `^` ở đầu chuỗi và `$` ở cuối chuỗi, mặc dù trên thực tế, bất kỳ ký tự riêng biệt nào không phải là ký tự viết thường hoặc dấu cách đều có thể hoạt động), và bao gồm ký tự đặc biệt đó trong chuỗi con mà bạn đang khớp với chuỗi đầu vào.

Ví dụ, nếu chúng ta muốn kiểm tra rằng `my_string` bắt đầu bằng `'at'`, chúng ta có thể sử dụng `in`, nhưng điều này cũng sẽ khớp với một chuỗi con ở giữa chuỗi, điều mà chúng ta không muốn. Thêm `^` vào đầu chuỗi cho phép chúng ta kiểm tra cụ thể `'^at'` mà sẽ chỉ khớp ở đầu chuỗi.

```python
my_string = "catty"
print('at' in my_string)  # đúng
```

```
True
```

```python
my_string = '^' + my_string
print('^at' in my_string)  # sai
```

```
False
```
