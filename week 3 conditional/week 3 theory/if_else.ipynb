# Câu lệnh If trong Python

Mảnh ghép cuối cùng trong câu đố cho phép chúng ta thực sự *thực hiện* những việc dựa trên một bài kiểm tra, đó là câu lệnh `if`:

```python
n = int(input("Nhập một số nguyên: "))
```

```python
if 0 < n < 6:
```

```python
    print('Bạn đã nhập một số nguyên dương nhỏ hơn sáu.')
```

```python
print('Hãy thử lại với một số nguyên khác!')
```

Đoạn mã trên lấy một số đầu vào vào biến `n`, sau đó kiểm tra điều kiện `0 < n < 6`. Nếu điều kiện đó được đánh giá là `True`, câu lệnh `print('Bạn đã nhập một số nguyên dương nhỏ hơn sáu.')` sẽ chạy. Nếu nó được đánh giá là `False` thì nó sẽ không chạy. Trong cả hai trường hợp, việc thực thi tiếp tục với `print('Hãy thử lại với một số nguyên khác!')`.

Điều này sẽ trông rất giống với khối if từ bài tập 0, hoàn chỉnh với mã bên trong câu lệnh `if` được *lồng* bên trong nó. Nghĩa là, câu lệnh `print()` đầu tiên ở đây được **thụt lề** (tức là có thêm khoảng trắng ở đầu dòng để bù đắp cho câu lệnh `if` mà nó được gắn vào). Đây là cách Python biết mã nào cần chạy khi điều kiện `if` được đánh giá là `True`. Nếu có nhiều dòng mã hơn được liên kết với câu lệnh `if`, chúng sẽ tương tự được thụt lề, với cùng một lượng chính xác như câu lệnh `print()`. Điều này tương tự như cách chúng ta có thể gắn thêm các khối mã bên dưới khối đầu tiên bên trong một khối if.

Mặc dù việc thụt lề về mặt kỹ thuật có thể là bao nhiêu khoảng trắng tùy thích, để tuân theo các thực hành phong cách Python tốt, bạn nên làm cho tất cả các lần thụt lề của mình dài **bốn khoảng trắng**. Grok sẽ tự động thêm bốn khoảng trắng thụt lề khi bạn nhấn phím "tab", như một phím tắt.

Tập hợp các câu lệnh thụt lề kết hợp bên dưới dòng `if` được gọi là một **khối**. Đừng nhầm lẫn, đây là một định nghĩa khác của "khối" so với các khối blockly mà chúng ta đã sử dụng để xây dựng chương trình trong bài tập 0. Dạng tổng quát của một câu lệnh `if` như sau:

```python
if <điều_kiện>:
    <khối_các_câu_lệnh>
```

`điều_kiện` thường là một biểu thức logic, liên quan đến các toán tử quan hệ hoặc logic. Khối các câu lệnh thụt lề chỉ được thực thi nếu điều kiện được đánh giá là `True`. Lưu ý rằng khái niệm "khối" này rất quan trọng trong Python, và sẽ xuất hiện lại trong bối cảnh của cả hàm và lặp.

## **"Thường"?**

Chúng ta đã nói ở trên rằng điều kiện trong câu lệnh `if` "**thường** là một biểu thức logic". Vậy, điều kiện **phải** là một biểu thức, vì nó phải được đánh giá thành một cái gì đó. Nhưng mặc dù việc nó là một biểu thức logic là tự nhiên nhất (khi chúng ta sử dụng `if` chúng ta thường kiểm tra một điều kiện nào đó để tạo ra kết quả chân lý) Python cho phép điều kiện được đánh giá thành bất kỳ kiểu dữ liệu nào bạn thích.

Nếu điều kiện không được đánh giá thành một `bool`, Python sẽ chuyển đổi nó một cách ngầm định thành một `bool`, sử dụng các quy tắc mà chúng ta đã thấy trước đây khi giới thiệu boolean. Nghĩa là, giá trị zero hoặc empty của một kiểu dữ liệu (`0`, `0.0`, `''`) sẽ được đánh giá thành `False` và mọi thứ khác là `True`.

Vậy,

```python
if "hello":
```

hoạt động, và tương đương với

```python
if bool("hello"):
```

có nghĩa là nó sẽ kích hoạt mỗi lần vì `bool('hello')` sẽ luôn được đánh giá thành `True`. Chúng ta sẽ thấy một số cách sử dụng thông minh hơn của điều này trong tương lai.

# Câu lệnh If-Else trong Python

Như chúng ta đã thấy trong bối cảnh của các chương trình turtle, nếu bạn cần quyết định giữa hai lựa chọn, Python cung cấp câu lệnh `if-else`:

```python
if <điều_kiện>:
    <khối lệnh đầu tiên>
else:
    <khối lệnh thay thế>
```

Nếu điều kiện được đánh giá là `True`, khối lệnh thụt lề đầu tiên sẽ được thực thi, ngược lại khối lệnh thay thế sẽ được thực thi. Tóm lại: câu lệnh `if-else` có thể mô hình hóa một quyết định hai chiều.

Hãy thử chạy ví dụ này:

```python
n = int(input("Nhập một số nguyên: "))

if 0 < n < 6:
    print('Bạn đã nhập một số nguyên dương nhỏ hơn sáu.')
else:
    print('Số nguyên không nằm trong khoảng từ không đến sáu.')
```

# Câu lệnh if đa điều kiện trong Python

Chúng ta nên làm gì nếu phải quyết định giữa nhiều lựa chọn, tức là nhiều hơn hai lựa chọn? Đối với trường hợp này, Python cung cấp cấu trúc quyết định đa hướng:

```python
if <điều_kiện_1>:
    <khối lệnh 1>
elif <điều_kiện_2>:
    <khối lệnh 2>
elif <điều_kiện_3>:
    <khối lệnh 3>
...
else:
    <khối lệnh n>
```

Hãy thử chạy ví dụ sau của câu lệnh `if` đa điều kiện và nhập các giá trị khác nhau để `in` ra các thông báo khác nhau:

```python
number = float(input("Nhập một số: "))

if number < 0:
    print(number, "là số âm")
elif number < 10:
    print(number, "là số nhỏ")
else:
    print(number, "là số lớn!")
```

Trong câu lệnh `if` đa điều kiện, **tối đa một** khối lệnh được thụt lề sẽ được thực thi. Các điều kiện được kiểm tra theo thứ tự cho đến khi tìm thấy điều kiện nào đó có giá trị `True`. Khối mã tương ứng sẽ được thực thi và tất cả các điều kiện và khối còn lại sẽ được bỏ qua. Nếu không có điều kiện nào là `True` nhưng có khối `else`, thì Python sẽ thực thi khối `else`.

## Lưu ý quan trọng:
- Chỉ **một khối duy nhất** sẽ được thực thi
- Các điều kiện được kiểm tra **theo thứ tự từ trên xuống dưới**
- Khi tìm thấy điều kiện đầu tiên là `True`, các điều kiện sau sẽ được **bỏ qua**
- Khối `else` chỉ chạy khi **tất cả điều kiện đều là `False`**

# Kết luận 

Vừa qua, chúng ta đã tìm hiểu về công cụ cuối cùng trong ba công cụ cần thiết để tạo ra code hoạt động khác nhau trong các tình huống khác nhau: một cách để chỉ chạy code khi một điều kiện nhất định được thỏa mãn.

* Câu lệnh `if` có thể được sử dụng để chạy một khối code chỉ khi một điều kiện cụ thể (thường là một biểu thức logic) được đánh giá là `True`.
* Các dòng code liên quan đến câu lệnh `if` được gọi là một **khối** (block).
* Để chỉ định rằng một khối được liên kết với một câu lệnh `if` cụ thể, mỗi dòng của khối đó phải được **thụt lề** bằng cùng một số khoảng trắng trên các dòng ngay bên dưới `if`. Trong môn học này, bạn nên luôn sử dụng bốn khoảng trắng để thụt lề.
* Một câu lệnh `else` có thể được bao gồm sau `if` để chỉ định một khối code thay thế để thực thi trong trường hợp điều kiện của `if` được đánh giá là `False`.
* Nhiều câu lệnh `elif` có thể được sử dụng sau `if` để chỉ định các kiểm tra thay thế sẽ được thử nếu tất cả các kiểm tra trước đó được đánh giá là `False`.
* Điều quan trọng là luôn xem xét việc bao gồm một mệnh đề `else` để tránh quên bất kỳ tình huống nào không được bao phủ bởi các điều kiện `if` và `elif` của bạn.

Dạng tổng quát của câu lệnh if như sau:

```python
if <điều kiện>:
    <khối code>
elif <điều kiện 2>:
    <khối code>
...
else:
    <khối code>
```
