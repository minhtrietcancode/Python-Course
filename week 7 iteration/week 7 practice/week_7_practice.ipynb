# Bài tập tạo từ

Trong bài tập này, bạn sẽ viết một chương trình in ra danh sách các từ có chung một hậu tố. Đầu tiên, chương trình của bạn phải yêu cầu người dùng nhập tất cả các từ **cách nhau bằng dấu cách**. Sau đó, chương trình phải yêu cầu nhập hậu tố chung được chia sẻ bởi mọi từ. Tạo các từ bằng cách lấy từng từ và thêm chuỗi thứ hai vào nó.

Ví dụ, nếu các từ ban đầu là `bake cake lake make rake take wake` và hậu tố là `s`, thì từ đầu tiên sẽ là `bakes` (`bake` + `s`), tiếp theo là `cakes`, `lakes`, `makes`, `rakes`, `takes`, và `wakes`.

Chương trình của bạn nên tạo ra tất cả các từ và in chúng ra từng dòng một. Sử dụng phương thức `.split()` để chuyển chuỗi các từ thành một danh sách và sau đó sử dụng vòng lặp `for` để truy cập từng từ. Kết quả đầu ra của chương trình sẽ trông như thế này:

# Bài toán Gamertag

Gamertag là một tên người dùng riêng biệt mà bạn sử dụng để định danh bản thân trong cộng đồng game trực tuyến. Với quy mô của những cộng đồng như vậy ngày càng lớn, rất có khả năng một tag đơn giản như tên của bạn có thể đã được sử dụng. Bạn vẫn có thể chọn sử dụng tên của mình làm tag nhưng với một số chỉnh sửa nhỏ như chèn dấu gạch ngang (`-`) sau mỗi chữ cái. Ví dụ, nếu tên của bạn là `Sandy`, thì gamertag sẽ là `S-a-n-d-y-`.

Viết một hàm `make_gamertag(name)` trả về gamertag khi được cho tên người dùng dưới dạng chuỗi. Hàm của bạn nên hoạt động như sau:

```python
>>> make_gamertag('Alex')
'A-l-e-x-'
```

# **Xác thực thẻ tín dụng**

Giả sử kết nối Internet tại cửa hàng bán lẻ của bạn tạm thời bị ngắt và bạn cần nhanh chóng xác thực số thẻ tín dụng của khách hàng trước khi gửi đi để ủy quyền ghi nợ. Đối với mục đích của chúng ta, một số thẻ tín dụng là hợp lệ nếu nó có 16 chữ số và tổng của tất cả các chữ số là bội số của 10, ngược lại thì không hợp lệ. Viết một hàm `validate_card(cardnum)` nhận một chuỗi các chữ số `cardnum` làm đầu vào (ví dụ: `'1111111111120133'`) và trả về `True` nếu `cardnum` là số thẻ tín dụng hợp lệ, và `False` trong trường hợp ngược lại.

Hàm của bạn nên hoạt động như thế này:

```python
print(validate_card('1111111111111133'))
```

```python
print(validate_card('1111111111111111'))
```

```python
print(validate_card('1111111111'))
```

**Kết quả mong đợi:**
```
True
```

```
False
```

```
False
```

Bạn có thể giả định rằng chỉ có các chữ số (số, không phải chữ cái) xuất hiện trong đối số đầu vào.

## **Gợi ý**
