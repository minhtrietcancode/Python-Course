# Cắt và Xử lý Chuỗi (String Slicing and Dicing)

Trong bài toán này, bạn phải viết một chương trình có thể trích xuất giá của, ví dụ, một thanh chocolate khi cho một câu như `How much did 25 chocolate bars cost? $63.75!` (25 thanh chocolate có giá bao nhiêu? $63.75!)

Chương trình của bạn phải:
1. Đầu tiên đọc một câu từ đầu vào
2. Sau đó sử dụng slicing để lấy số lượng sản phẩm và tổng chi phí
3. Chuyển đổi các giá trị đó từ chuỗi thành số
4. Cuối cùng, tính toán và in ra giá của một sản phẩm đơn lẻ

Đầu ra của chương trình sẽ trông như thế này:

```
Question: How much did 25 chocolate bars cost? $63.75!
```

```
$2.55
```

Chương trình của bạn vẫn phải hoạt động nếu câu chứa số lượng, giá hoặc loại sản phẩm khác nhau:

```
Question: How much did 10 cans of cola cost? $35.00!
```

```
$3.50
```

## Giả định cho bài toán này:

- **Số lượng sản phẩm luôn là 2 chữ số** (tức là từ `10` đến `99`) và xuất hiện sau cụm từ `'How much did '`
- **Tổng chi phí luôn từ 10 đến 99 đô la có kèm xu** (dưới dạng hai chữ số), và xuất hiện ở cuối câu (theo sau bởi dấu chấm than)

## Gợi ý

Bạn có thể muốn chuyển đổi số tiền từ đô la và xu thành tổng số xu, và sử dụng phép chia nguyên (`//`) và phép chia lấy dư (`%`) để tách riêng số đô la và số xu.

# A và An

Trong tiếng Anh, mạo từ bất định có hai dạng: *a* (được sử dụng khi âm đầu tiên của từ tiếp theo là phụ âm) và *an* (được sử dụng, nói một cách đại khái, khi âm đầu tiên của từ tiếp theo là nguyên âm). Để đơn giản hóa cách phân biệt này, chúng ta sẽ giả định rằng *an* nên được sử dụng khi chữ cái đầu tiên của từ đứng trước nó là một trong các chữ *a*, *e*, *i*, *o* và *u*, và *a* nên được sử dụng trong các trường hợp khác.

Viết một chương trình yêu cầu người dùng nhập một cụm từ, và in ra cụm từ đó được đi trước bởi `an` (với một khoảng trắng) nếu cụm từ bắt đầu bằng nguyên âm, và `a` (với một khoảng trắng) cho tất cả các đầu vào khác **(bao gồm cả chuỗi rỗng)**.

Chương trình của bạn nên hoạt động như sau:

```
Enter a phrase: aardvark
```

```
an aardvark
```

```
Enter a phrase: banana
```

```
a banana
```

```
Enter a phrase: Australian
```

```
an Australian
```

Lưu ý rằng, dựa trên phương pháp đơn giản của chúng ta, phương pháp này sẽ tạo ra một số đầu ra nghe có vẻ khó hiểu hơn, chẳng hạn như:

```
Enter a phrase: honour
```

```
a honour
```

```
Enter a phrase: unicorn
```

```
an unicorn
```

# Kẻ Thủ Phạm

James Bond đã bắt giữ 7 kẻ thù không đội trời chung của mình. Anh ta biết rằng một trong số họ là thủ phạm đằng sau kế hoạch độc ác nhất trong vũ trụ, nhưng không chắc chắn đó là ai. Tuy nhiên, anh ta có một mánh khóe. Anh ta biết rằng Hugo (một trong 7 người) không quá thông minh và Hugo cũng biết ai đã làm điều đó. Vì vậy, anh ta sẽ đưa cho Hugo một danh sách tên với một số thông tin gây nhầm lẫn bên cạnh, rồi nhanh chóng hỏi Hugo xem ai đã làm điều đó. Hugo đủ thông minh để không nói chính xác ai đã làm, thay vào đó anh ta sẽ luôn nói tên của người *đứng trước* trong danh sách, danh sách này được đánh số từ 1 đến 7 (với câu trả lời là 1 có nghĩa là người thứ 2, và 7 có nghĩa là thủ phạm là người thứ 1).

Viết một chương trình yêu cầu Hugo đưa ra số của thủ phạm, và trả về tên của người đã làm điều đó cùng với dữ liệu liên quan đến cá nhân đó (tức là mọi thứ đi sau tên trong tuple). Danh sách các thủ phạm có thể có nên được lưu trữ dưới dạng một danh sách các tuple, như bên dưới, trong đó mỗi tuple chứa một tên (một chuỗi) và một số lượng biến đổi các điểm dữ liệu liên quan đến cá nhân đó (mỗi điểm là một chuỗi). Giả sử đầu vào là một số nguyên dương. Dữ liệu liên quan đến cá nhân phải được trả về dưới dạng một tuple (có thể rỗng).

## **Độ lệch zero và danh sách lồng nhau**

Hãy nhớ rằng danh sách được lập chỉ mục từ 0. Ngoài ra, bạn có thể thấy việc lập chỉ mục **lồng nhau** hữu ích ở đây.

```python
my_list = [(4, 8)]
print(my_list[0][0])  # In ra: 4
print(my_list[0][1])  # In ra: 8
```
