# **Ôn tập: Số vs Dãy**

Như Python Data Model nêu rõ, mỗi đối tượng có một kiểu dữ liệu, và kiểu dữ liệu đó ảnh hưởng đến những gì chúng ta có thể làm với từng phần dữ liệu. Vậy hãy cùng xem xét hai loại kiểu dữ liệu chúng ta đã thấy cho đến nay: **Số** và **Dãy**.

## **Số**

| **Kiểu** | **Ví dụ** | **Lưu trữ gì** | **Một số cách sử dụng** |
|-----------|-----------|----------------|-------------------------|
| `int` | `-3, -5, 10` | Số nguyên | Đếm, đánh số, chỉ định chỉ mục và cắt lát |
| `float` | `-3.0, 0.5, 3.14159` | Số thực (có phần thập phân) | Biểu diễn số thực cho toán học và đo lường |
| `bool` | `True, False` | Giá trị logic nhị phân | Kết quả của kiểm tra logic/biểu thức logic, hữu ích với câu lệnh điều kiện `if` |

**Toán tử:**
* Toán tử số học: `*`, `/`, `**`, `//`, `%`, `+`, `-`. Toán tử một ngôi `+` và `-`.
* Toán tử quan hệ: `<`, `<=`, `==`, `!=`, `>`, `>=`
* Toán tử logic: `and`, `or`, `not`

Ngạc nhiên khi thấy Boolean ở đây như một kiểu số? Bool về mặt kỹ thuật cũng là số, và hoạt động như `1` và `0` tương ứng với `True` và `False`. Hãy thử chạy `True + True` nếu bạn không tin!

Các toán tử logic chỉ có ý nghĩa khi áp dụng cho giá trị `bool`, mặc dù chúng có thể, một cách khó hiểu, hoạt động với bất kỳ kiểu nào. Thay vì đi sâu vào tại sao `not 'hello' or 'hi'` cho kết quả `'hi'`, hãy đồng ý chỉ sử dụng toán tử logic cho giá trị boolean.

## **Dãy**

| **Kiểu** | **Ví dụ** | **Lưu trữ gì** | **Một số cách sử dụng** |
|-----------|-----------|----------------|-------------------------|
| `str` | `"Hello"` | Dãy ký tự | Biểu diễn văn bản, nhận đầu vào, in đầu ra. Các phương thức như `.upper()` |
| `list` | `[1, 3.3, "h"]` | Dãy bất kỳ thứ gì | Lưu trữ tập hợp (độ dài thay đổi) các đối tượng |
| `tuple` | `(1.5, 2, "h")` | Dãy không thể chỉnh sửa bất kỳ thứ gì | Lưu trữ tập hợp (độ dài cố định) các đối tượng |

**Toán tử:**
* `+` (nối chuỗi), `*` (lặp lại)
* Toán tử quan hệ: `<`, `<=`, `==`, `!=`, `>`, `>=`, bao gồm `in`

`len()` có thể được sử dụng để tìm độ dài của bất kỳ dãy nào.

Các phần tử có thể được trích xuất từ dãy bằng cách **lập chỉ mục**. Điều này được thực hiện bằng cách đặt chỉ mục của phần tử bạn muốn trong dấu ngoặc vuông sau đối tượng: `(1, 2, 3)[1]`. Nhớ rằng việc đếm bắt đầu từ `0` từ trái sang phải hoặc `-1` từ phải sang trái!

Tất cả các dãy có thể được **cắt lát** bằng cách đặt hai chỉ mục trong dấu ngoặc vuông, cách nhau bởi dấu hai chấm. Số thứ ba tùy chọn điều khiển bước nhảy. Chỉ mục có thể được bỏ qua để cắt đến cuối dãy: `(1, 2, 3)[:1]`

Chuỗi có thể được định dạng bằng `f`-strings và bộ chỉ định định dạng. Điều này là duy nhất đối với chuỗi: đừng thử viết `f`-lists!

Một điểm thú vị là `print()` hoạt động bằng cách chuyển đổi các đối số của nó thành chuỗi để hiển thị chúng trên màn hình. Chỉ là một lý do khác tại sao chuỗi rất hữu ích.

# Ôn tập: Chuyển đổi kiểu dữ liệu

Mặc dù chúng ta đã thấy **chuyển đổi kiểu dữ liệu** (còn gọi là **ép kiểu**) từng trường hợp cụ thể, đây là một phép toán tổng quát: bạn thường có thể chuyển đổi giữa các kiểu dữ liệu (khi việc này có ý nghĩa) bằng cách sử dụng một hàm có cùng tên với kiểu dữ liệu đó. Xem các ví dụ quan trọng dưới đây:

```python
# str sang int
print(int("4"))
```

```python
# int sang str
print(str(4+3))
```

```python
# int sang float
print(float(4))
```

```python
# str sang float
print(float("4"))
```

```python
# str sang list
print(list("hello"))
```

```python
# tuple sang list
print(list(("me1","me2")))
```

```python
# str sang bool
print(bool("hello"))
```

Chúng ta đã xem xét các kiểu dữ liệu mà các phép toán cơ bản trả về:
* Khi các toán hạng của phép cộng, trừ hoặc nhân đều là `int`, kết quả sẽ luôn là một `int`.
* Khi ít nhất một toán hạng của `+`, `-` hoặc `*` là `float`, kết quả sẽ luôn là một `float`.
* Khi sử dụng phép chia, kết quả sẽ luôn là `float` bất kể các toán hạng có kiểu `int` hay `float`.

**Nhớ về Chuyển đổi Boolean**

Các chuyển đổi cần nhớ là chuyển đổi kiểu Boolean vì chúng được thực hiện ngầm định nếu bạn có điều kiện trên một giá trị. Ví dụ, `if "hello"` giống như `if bool("hello")`. Liệu đoạn mã dưới đây có in ra `hello` không?

```python
if "hello":
    print("hello")
```

```python
print(bool("hello"))
```

# **Ôn tập: Câu lệnh và Biểu thức**

Về mặt ngữ pháp của ngôn ngữ lập trình Python, mỗi dòng code có thể được chia thành các câu lệnh và biểu thức. Chúng ta đã tìm hiểu các thuật ngữ sau:

**Biểu thức (Expression)**: một đoạn code tạo ra một đối tượng. Đây có thể là một literal hoặc một identifier đã được gán trước đó. Nó có thể là một phép toán, hoặc một lời gọi hàm/phương thức. Nó có thể là một chỉ số hoặc slice của một chuỗi. Một lời gọi hàm có thể được coi như một biểu thức được đánh giá thành giá trị trả về của nó.

**Biểu thức đơn giản (Simple Expression)**: Một biểu thức chỉ chứa một toán tử và các toán hạng của nó dưới dạng literal hoặc biến.

**Biểu thức phức tạp (Complex Expression)**: Một biểu thức có nhiều phần hơn so với biểu thức đơn giản. Đây có thể là biểu thức với một toán tử mà các toán hạng chính là các biểu thức khác, chẳng hạn như các toán hạng của `/` trong `(2 + 4) / (4 - 3)`

**Biểu thức logic (Logical Expression)**: Một biểu thức được đánh giá thành `bool`. Thường chứa một số kết hợp của các toán tử logic và quan hệ.

**Câu lệnh (Statement)**: Bất cứ thứ gì là một dòng code Python hợp lệ. Điều này bao gồm tất cả các biểu thức, nhưng cũng bao gồm các dòng với `def`, `return`, `if`, `elif`, `else`. Nó cũng bao gồm các phép gán (`=`).

Phù! Hy vọng chúng tôi không làm bạn quá bối rối với quá nhiều thuật ngữ. Trong thực tế, miễn là chúng ta biết cách *sử dụng* những khái niệm này trong code của mình, việc có thể phân biệt một số thuật ngữ tinh tế này trong mô tả code bằng ngôn ngữ thông thường ít quan trọng hơn ở thời điểm này.

Chỉ để tóm tắt slide này, dưới đây là một số đoạn code để minh họa những sự phân biệt tinh tế mà chúng tôi đang cố gắng thực hiện (nhưng một lần nữa, hãy thoải mái bỏ qua điều này như quá cầu kỳ ở thời điểm này nếu bạn bị lạc trong các chi tiết!).

# **Ôn tập: Câu lệnh có dấu hai chấm**

Chúng ta đã học rằng khi có dấu hai chấm ở cuối một dòng code, dòng tiếp theo phải được thụt lề. Bốn dấu cách là số lượng thụt lề chuẩn trong môn học này. Các cấu trúc code mà chúng ta đã thấy quy tắc này bao gồm:

## **Câu lệnh điều kiện**

Chúng ta đã thấy cách có thể kiểm soát luồng thực thi bằng cách chọn giữa một số nhánh để theo bằng câu lệnh `if`. Sau `if`, Python cho phép bao nhiêu mệnh đề `elif` tùy ý (hoặc ít, bao gồm khả năng không có), theo sau bởi một câu lệnh `else` tùy chọn.

```python
if <điều_kiện>:
    <khối_code>
elif <điều_kiện_2>:
    <khối_code_2>
...
else:
    <khối_code_n>

<việc_thực_thi_tiếp_tục_tại_đây>
```

Python bắt đầu với `điều_kiện` đầu tiên, đánh giá nó (chuyển đổi ngầm thành `bool` nếu nó không đánh giá thành boolean) và nếu nó là `True`, khối code đầu tiên được chạy, với việc thực thi sau đó tiếp tục sau khi kết thúc cấu trúc điều kiện. Mỗi điều kiện `elif` sau đó được kiểm tra theo cách tương tự như `if` đầu tiên. Nếu không có điều kiện `if` và `elif` nào trả về `True`, khối `else` sẽ được chạy.

Hãy nhớ rằng tối đa một khối sẽ được chạy mỗi khi đạt đến một cấu trúc điều kiện. Nếu nhiều hơn một điều kiện đúng, chỉ có điều kiện đầu tiên (theo thứ tự) sẽ được chạy. Nếu không có mệnh đề `else`, không có khối nào sẽ được chạy nếu tất cả các điều kiện đánh giá thành `False`.

## **Hàm**

Chúng ta đã thấy cách định nghĩa các hàm riêng cho các tác vụ phổ biến, có thể được gọi trong code với các giá trị tham số đầu vào khác nhau. Hàm là một cách tốt để làm cho code của chúng ta ít lặp lại hơn và được tổ chức tốt hơn.

```python
def <tên_hàm>(<tham_số>):
    <thân_hàm>
```

Nhiều tham số có thể được chỉ định bằng cách phân tách chúng bằng dấu phẩy; một hàm không có tham số cũng hợp lệ. Hàm có thể được gọi bằng cách viết tên của nó và đặt các tham số trong ngoặc đơn. Thân hàm sẽ được chạy, và nếu một câu lệnh `return` được đạt đến, giá trị trả về sẽ được truyền lại.

Chúng ta cũng đã xem xét các tham số có thể được đặt tên hoặc được gán giá trị mặc định.
