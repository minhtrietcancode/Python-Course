# **Bài Toán Cờ Vua v1**

Viết một hàm `check_move(column, row)` trả về một chuỗi mô tả một nước đi cờ vua đến một hàng và cột nhất định trên bàn cờ. Chương trình của bạn phải kiểm tra xem hàng và cột đã nhập có hợp lệ hay không. Cột trong bàn cờ vua là một chữ cái từ **A đến H** (bao gồm, viết hoa) và hàng là một số từ **1 đến 8** (bao gồm).

* `column` sẽ được truyền vào hàm của bạn dưới dạng chuỗi và `row` sẽ được truyền dưới dạng số nguyên.
* `'A'` hoặc `'E'` là các cột hợp lệ nhưng `'a'` hoặc `'L'` thì không (viết thường và nằm ngoài phạm vi, tương ứng).
* Tương tự, `2` và `8` là các hàng hợp lệ nhưng `0` và `9` thì không (cả hai đều nằm ngoài phạm vi).

Nếu cả hai tọa độ đều hợp lệ, chẳng hạn như `'E'` và `2`, hàm của bạn phải trả về `'The piece is moved to E2.'`, ngược lại nó sẽ trả về `'The position is not valid.'`.

Lưu ý rằng bạn có thể giả định rằng đối số đầu tiên (ký hiệu cột) là một `str` có một ký tự, và đối số thứ hai (ký hiệu hàng) là một `int`.

Ví dụ, `check_move` sẽ hoạt động như thế này với các đối số hợp lệ:

```python
>>> check_move('B', 4)
```

```
'The piece is moved to B4.'
```

và như thế này với các đối số không hợp lệ:

```python
>>> check_move('b', 4)
```

```
'The position is not valid.'
```

```python
>>> check_move('I', 4)
```

```
'The position is not valid.'
```

```python
>>> check_move('F', 9)
```

```
'The position is not valid.'
```

# **Bài toán Cờ vua v4**

OK, điểm dừng cuối cùng trong thế giới cờ vua, mở rộng hàm `check_move()` một lần nữa. Trước đây, hàm của bạn nhận hai tham số riêng biệt: giá trị cột và giá trị hàng. Bây giờ bạn sẽ viết lại hàm để chấp nhận vị trí bàn cờ như một tham số `str` duy nhất. Nói cách khác, đầu vào của `check_move(position)` giờ sẽ là một chuỗi vị trí duy nhất như `'B5'`, mã hóa cả ký hiệu cột và hàng.

* Khi cả hai tọa độ trong `position` đều hợp lệ, ví dụ `'c4'`, hàm trả về `'The piece is moved to C4.'`.
* Nếu `position` có quá nhiều hoặc quá ít ký tự, trả về `'The position is not valid.'`
* Nếu ký hiệu cột nằm ngoài phạm vi (bất kể ký hiệu hàng có hợp lệ hay không), trả về `'The column value is not in the range a-h or A-H!'`.
* Ngược lại, nếu ký hiệu hàng nằm ngoài phạm vi, trả về `'The row value is not in the range 1 to 8!'`.

Lưu ý rằng, bất kể việc viết hoa hay thường của giá trị cột, mã của bạn nên xuất ra giá trị viết hoa. Cũng lưu ý rằng không có đảm bảo rằng đầu vào sẽ có đúng hai ký tự.

# Bài toán Mã Bảo Mật Thẻ

Mã bảo mật thẻ 3 chữ số (CSC) trên thẻ tín dụng giúp bảo vệ chống lại gian lận thẻ tín dụng. Viết một hàm đơn giản `check_csc(code)` để kiểm tra xem một mã cho trước (được nhập dưới dạng chuỗi) có hợp lệ hay không. Đối với mục đích của chúng ta, mã này hợp lệ nếu nó có **chính xác ba ký tự** và cả ba ký tự đều là **chữ số từ 0 đến 9 (bao gồm cả 0 và 9)**. Nếu CSC hợp lệ, hàm của bạn phải trả về `True`. Nếu không, nó phải trả về `False`.

Hàm của bạn sẽ hoạt động như sau:

```python
>>> check_csc('023')
True
```

```python
>>> check_csc('23')
False
```

```python
>>> check_csc('Ab3')
False
```

```python
>>> check_csc('')
False
```
