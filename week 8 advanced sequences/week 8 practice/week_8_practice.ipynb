# Cycling Lists (Xoay vòng danh sách)

Viết một hàm `cycle(input_list)` thực hiện việc "xoay vòng tại chỗ" các phần tử của một danh sách, di chuyển mỗi phần tử lên một vị trí trước đó trong danh sách. Ở đây "tại chỗ" có nghĩa là thao tác này phải được thực hiện bằng cách biến đổi danh sách gốc, và hàm của bạn không nên trả về gì cả. Lưu ý rằng bạn có thể giả định rằng `input_list` không rỗng (tức là chứa ít nhất một phần tử).

Ví dụ:

```python
>>> my_list = [1, 2, 4, 5, 'd']
```

```python
>>> cycle(my_list)
```

```python
>>> my_list
```

```python
[2, 4, 5, 'd', 1]
```

```python
>>> cycle(my_list)
```

```python
>>> my_list
```

```python
[4, 5, 'd', 1, 2]
```

Viết một hàm `sorted_words(wordlist)` nhận một đối số duy nhất là danh sách từ `wordlist`, và trả về một danh sách đã sắp xếp các từ trong `wordlist` mà các chữ cái được sắp xếp theo thứ tự bảng chữ cái. Một ví dụ về từ như vậy là `door`, vì không có chữ cái nào trong từ có giá trị Unicode cao hơn bất kỳ chữ cái nào theo sau nó, trong khi `cat` thì không, vì `c` đứng trước `a` trong từ (gợi ý: hàm `sorted()` có thể hữu ích trong việc kiểm tra xem các chữ cái trong một từ có được sắp xếp theo thứ tự bảng chữ cái hay không).
```python
>>> sorted_words(["bet", "abacus", "act", "celebration", "door"])
```


```python
>>> sorted_words(['apples', 'bananas', 'spam'])
```

```
[]
```

```python
>>> sorted_words(["aims", "Zip"])

```

# Tìm (Các) Từ Ở Giữa

Hãy tưởng tượng chúng ta có danh sách các từ sau:

```python
['Mirror', 'Mirror', 'on', 'the', 'wall']
```

Có 5 từ trong danh sách này, với từ ở giữa là `'on'`. Khi độ dài của danh sách là một số lẻ, thì có chính xác một từ ở giữa. Mặt khác, nếu độ dài là số chẵn, thì có hai từ ở giữa. Ví dụ, trong trường hợp:

```python
['Baa', 'baa', 'black', 'sheep', 'have', 'you', 'any', 'wool']
```

hai từ ở giữa là `'sheep'` và `'have'`.

Nhiệm vụ của bạn là viết một hàm `middle_words(word_list)` trả về (các) từ ở giữa từ danh sách các chuỗi không rỗng `word_list`. Nếu độ dài của `word_list` là một số lẻ, bạn nên trả về một danh sách chứa từ ở giữa duy nhất; và nếu độ dài của `word_list` là một số chẵn, bạn nên trả về một danh sách chứa hai từ ở giữa, theo cùng thứ tự như chúng xuất hiện trong `word_list`. 

Ví dụ:

```python
>>> middle_words(['Mirror', 'Mirror', 'on', 'the', 'wall'])
["on"]
```

```python
>>> middle_words(['Baa', 'baa', 'black', 'sheep', 'have', 'you', 'any', 'wool'])
["sheep", "have"]
```

# Câu Dài Nhất

Như bạn đã thấy trước đó, phương thức `.split()` sẽ chia một chuỗi thành các từ riêng biệt, dựa trên hành vi mặc định là chia chuỗi tại mỗi lần xuất hiện của khoảng trắng.

Bạn cũng có thể sử dụng `.split()` để chia chuỗi theo các ký tự khác nhau. Ví dụ, bạn có thể chia chuỗi theo dấu chấm bằng cách sử dụng `.split('.')`. Điều này cung cấp cho bạn một cách dễ dàng (mặc dù có phần đơn giản) để tạo ra tất cả các câu trong một chuỗi:

```python
my_string = 'Hello world. My name is Plargleflarp.'
```

```python
print(my_string.split('.'))
```

```
['Hello world', ' My name is Plargleflarp', '']
```

Viết một hàm `longest_sentence_length(text)` nhận một tham số chuỗi `text` và trả về **độ dài** của câu dài nhất trong `text`, được đo bằng số từ. Ví dụ:

```python
>>> text = 'Hello world. My name is Plargleflarp.'
```

```python
>>> longest_sentence_length(text)
```

```
4
```

Các bài kiểm tra cho bài tập này sử dụng đoạn văn đầu tiên từ *Moby Dick* đã thấy trong bài toán "Độ Dài Từ Trước" ở phần trước trong worksheet này. Bạn có thể muốn sử dụng văn bản đó để kiểm tra hàm của mình ở đây.

## `.split()` với khoảng trắng

Nếu không có tham số nào, `.split()` sẽ chia dựa trên bất kỳ khoảng trắng nào, bao gồm cả dấu cách. Tuy nhiên, nếu chúng ta truyền ký tự dấu cách cho `.split(' ')` để làm cho nó hoàn toàn cụ thể cho dấu cách, nó hoạt động hơi khác khi có dấu cách ở đầu. Hãy lưu ý điều này khi viết code giải pháp cho bài toán này.

```python
my_string = ' hello'
```

```python
print(my_string.split())
```

```python
print(my_string.split(' '))
```

Chúng tôi đảm bảo sẽ không có ký tự khoảng trắng đặc biệt nào trong các trường hợp kiểm tra ngoài dấu cách.
