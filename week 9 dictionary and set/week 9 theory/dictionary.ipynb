# **L·∫≠p ch·ªâ m·ª•c cho Dictionary (T·ª´ ƒëi·ªÉn)**

M·ªôt key nh·∫•t ƒë·ªãnh ch·ªâ c√≥ th·ªÉ xu·∫•t hi·ªán m·ªôt l·∫ßn trong dictionary v√† ƒë∆∞·ª£c li√™n k·∫øt v·ªõi m·ªôt gi√° tr·ªã duy nh·∫•t (nh∆∞ng t·∫•t nhi√™n, b·∫°n c√≥ th·ªÉ l√†m cho gi√° tr·ªã ƒë√≥ l√† m·ªôt list ch·ª©a nhi·ªÅu ƒë·ªëi t∆∞·ª£ng). # C√°c Ph∆∞∆°ng Th·ª©c Dictionary

H√£y c√πng t√¨m hi·ªÉu m·ªôt s·ªë ph∆∞∆°ng th·ª©c c·ªßa dictionary.

Ph∆∞∆°ng th·ª©c `.get()` nh·∫≠n m·ªôt key l√†m tham s·ªë v√† tr·∫£ v·ªÅ gi√° tr·ªã li√™n k·∫øt v·ªõi n√≥. S·ª± kh√°c bi·ªát gi·ªØa ph∆∞∆°ng th·ª©c n√†y v√† vi·ªác truy xu·∫•t tr·ª±c ti·∫øp l√† trong tr∆∞·ªùng h·ª£p key kh√¥ng t·ªìn t·∫°i, vi·ªác truy xu·∫•t tr·ª±c ti·∫øp s·∫Ω n√©m ra l·ªói trong khi s·ª≠ d·ª•ng `.get()` s·∫Ω ƒë∆°n gi·∫£n tr·∫£ v·ªÅ `None`.

Ph∆∞∆°ng th·ª©c `.pop()` nh·∫≠n m·ªôt key l√†m tham s·ªë v√† tr·∫£ v·ªÅ gi√° tr·ªã, ƒë·ªìng th·ªùi **x√≥a** c·∫∑p key-value ƒë√≥. ƒêi·ªÅu n√†y t∆∞∆°ng t·ª± nh∆∞ c√°ch `.pop()` ƒë∆∞·ª£c s·ª≠ d·ª•ng cho lists ·ªü ch·ªó n√≥ thay ƒë·ªïi dictionary ƒë∆∞·ª£c g·ªçi v√† tr·∫£ v·ªÅ gi√° tr·ªã n·∫øu th·ª±c thi th√†nh c√¥ng, ho·∫∑c `KeyError` n·∫øu key kh√¥ng ƒë∆∞·ª£c t√¨m th·∫•y trong dictionary (lo·∫°i l·ªói n√†o ƒë∆∞·ª£c n√©m ra b·ªüi `.pop()` trong tr∆∞·ªùng h·ª£p c·ªßa list khi n√≥ th·∫•t b·∫°i?).

Cu·ªëi c√πng, ph∆∞∆°ng th·ª©c `.clear()` x√≥a to√†n b·ªô n·ªôi dung c·ªßa m·ªôt dictionary. Gi·ªëng nh∆∞ `.pop()`, ƒë√¢y l√† m·ªôt ph∆∞∆°ng th·ª©c thay ƒë·ªïi, v√¨ v·∫≠y ƒë·ª´ng g√°n n√≥ cho b·∫•t c·ª© th·ª© g√¨!

# C·∫≠p nh·∫≠t Dictionary

Trong tr∆∞·ªùng h·ª£p b·∫°n ƒë√£ b·ªè l·ª° t·ª´ slide tr∆∞·ªõc, dictionary l√† **c√≥ th·ªÉ thay ƒë·ªïi** (mutable). V√¨ v·∫≠y, gi·ªëng nh∆∞ list, n·ªôi dung c·ªßa ch√∫ng c√≥ th·ªÉ ƒë∆∞·ª£c thay ƒë·ªïi sau khi ƒë√£ ƒë∆∞·ª£c kh·ªüi t·∫°o. B·∫°n c√≥ th·ªÉ thay ƒë·ªïi gi√° tr·ªã ƒë∆∞·ª£c li√™n k·∫øt v·ªõi m·ªôt key ho·∫∑c th√™m m·ªôt c·∫∑p key-value m·ªõi v√†o dictionary b·∫±ng c√°ch s·ª≠ d·ª•ng to√°n t·ª≠ g√°n (`=`):

```python
capitals = {'Victoria': 'Melbie',
            'New South Wales': 'Sydney',
            'Queensland': 'Brisbane',
            'Tasmania': 'Hobart',
            'South Australia': 'Adelaide',
            'Western Australia': 'Perth'}

print(capitals['Victoria'])

capitals['Victoria'] = 'Melbourne'
capitals['ACT'] = 'Canberra'

print(capitals['Victoria'])
print(capitals['ACT'])
```

`'Victoria'` b√¢y gi·ªù √°nh x·∫° ƒë·∫øn gi√° tr·ªã `'Melbourne'` thay v√¨ `'Melbie'`, do ƒë√≥ s·ª≠a l·ªói tr∆∞·ªõc ƒë√≥ (ph√π!). Key `'ACT'` kh√¥ng c√≥ trong dictionary tr∆∞·ªõc c√¢u l·ªánh g√°n, v√¨ v·∫≠y ƒëo·∫°n code tr√™n m·ªü r·ªông dictionary v·ªõi m·ªôt m·ª•c √°nh x·∫° m·ªõi.

# Ki·ªÉm tra th√†nh vi√™n trong Dictionary

B·∫°n c√≥ th·ªÉ ki·ªÉm tra xem m·ªôt key c√≥ t·ªìn t·∫°i trong dictionary hay kh√¥ng b·∫±ng c√°ch s·ª≠ d·ª•ng to√°n t·ª≠ `in`. H√£y th·ª≠ ch·∫°y ƒëo·∫°n code sau v√†i l·∫ßn, nh·∫≠p c·∫£ t√™n bang h·ª£p l·ªá v√† kh√¥ng h·ª£p l·ªá:

```python
capitals = {'Victoria': 'Melbourne',
            'New South Wales': 'Sydney',
            'Queensland': 'Brisbane',
            'Tasmania': 'Hobart',
            'South Australia': 'Adelaide',
            'Western Australia': 'Perth'}

state = input('Nh·∫≠p t√™n bang: ')

if state in capitals:
    print(capitals[state])
else:
    print(state, 'kh√¥ng t√¨m th·∫•y')
```

To√°n t·ª≠ `in` ƒë·∫∑c bi·ªát h·ªØu √≠ch v√¨ n·∫øu ch√∫ng ta ch·∫°y d√≤ng code `print(capitals[state])` m√† gi√° tr·ªã c·ªßa `state` kh√¥ng t·ªìn t·∫°i nh∆∞ m·ªôt key trong dictionary, n√≥ s·∫Ω t·∫°o ra l·ªói `KeyError` nh∆∞ ƒë∆∞·ª£c hi·ªÉn th·ªã b√™n d∆∞·ªõi. V·ªõi `in`, ch√∫ng ta c√≥ th·ªÉ x√°c nh·∫≠n r·∫±ng m·ªôt key t·ªìn t·∫°i v√† s·∫Ω kh√¥ng g√¢y ra l·ªói tr∆∞·ªõc khi s·ª≠ d·ª•ng n√≥.

# Truy c·∫≠p t·∫•t c·∫£ c√°c Keys (Kh√≥a)

B·∫°n c√≥ th·ªÉ l·∫•y c√°c kh√≥a trong dictionary b·∫±ng ph∆∞∆°ng th·ª©c `.keys()`. `.keys()` tr·∫£ v·ªÅ m·ªôt collection ƒë·∫∑c bi·ªát c√≥ th·ªÉ l·∫∑p ƒë∆∞·ª£c g·ªçi l√† `dict_keys` view object, h·ªó tr·ª£ vi·ªác l·∫∑p v√† to√°n t·ª≠ `in` gi·ªëng nh∆∞ `list`, nh∆∞ng **kh√¥ng** h·ªó tr·ª£ indexing. N·∫øu b·∫°n mu·ªën index, b·∫°n s·∫Ω ph·∫£i chuy·ªÉn ƒë·ªïi n√≥ th√†nh `list()` m·ªôt c√°ch r√µ r√†ng.

```python
capitals = {'Victoria': 'Melbourne',
            'New South Wales': 'Sydney'}

keys = capitals.keys()

print("The keys:", keys)

print("Looping through all keys:")
for key in keys:
    print(key)

state = 'Victoria'
print(f"Is '{state}' in keys?", state in keys)

print(keys[0])  # ƒê√¢y l√† L·ªñI!
```

**Output:**
```
The keys: dict_keys(['New South Wales', 'Victoria'])
Looping through all keys:
New South Wales
Victoria
Is 'Victoria' in keys? True
Traceback (most recent call last):
  File "program.py", line 12, in <module>
    print(keys[0])  # This is an ERROR!
TypeError: 'dict_keys' object does not support indexing
```

B·∫°n c√≥ th·ªÉ th·∫Øc m·∫Øc "*view object* l√† g√¨?" Kh√¥ng c·∫ßn ph·∫£i lo l·∫Øng v·ªÅ c√°c chi ti·∫øt ph·ª©c t·∫°p; t·∫•t c·∫£ nh·ªØng g√¨ b·∫°n c·∫ßn bi·∫øt l√† ch√∫ng h·ªó tr·ª£ m·ªôt t·∫≠p con c√°c ch·ª©c nƒÉng m√† list h·ªó tr·ª£ - l·∫∑p v√† ki·ªÉm tra th√†nh vi√™n. ƒê·ªëi v·ªõi b·∫•t k·ª≥ h√†nh vi n√†o kh√°c m√† b·∫°n c√≥ th·ªÉ mu·ªën, b·∫°n c√≥ th·ªÉ chuy·ªÉn ƒë·ªïi c√°c view object th√†nh list b·∫±ng c√°ch s·ª≠ d·ª•ng h√†m chuy·ªÉn ƒë·ªïi `list()` tr∆∞·ªõc.

## C√≥ h·ªØu √≠ch kh√¥ng?

Ph∆∞∆°ng th·ª©c `.keys()` kh√¥ng th·ª±c s·ª± h·ªØu √≠ch trong th·ª±c t·∫ø, b·ªüi v√¨ vi·ªác tham chi·∫øu ƒë·∫øn m·ªôt dictionary s·∫Ω t·ª± ƒë·ªông tham chi·∫øu ƒë·∫øn c√°c kh√≥a c·ªßa n√≥.

Ch√∫ng ta ƒë√£ th·∫•y ·ªü slide tr∆∞·ªõc ƒë√≥ r·∫±ng to√°n t·ª≠ `in` ki·ªÉm tra th√†nh vi√™n trong c√°c kh√≥a c·ªßa dictionary. N·∫øu ch√∫ng ta mu·ªën l·∫∑p qua c√°c kh√≥a c·ªßa dictionary nh∆∞ tr√™n, ch√∫ng ta c√≥ th·ªÉ ch·ªâ c·∫ßn l·∫∑p qua b·∫£n th√¢n dictionary:

```python
capitals = {'Victoria': 'Melbourne',
            'New South Wales': 'Sydney'}

for key in capitals:
    print(key)
```

# Truy c·∫≠p t·∫•t c·∫£ Keys v√† Values

M·ªôt ph∆∞∆°ng th·ª©c h·ªØu √≠ch kh√°c cho dictionary l√† ph∆∞∆°ng th·ª©c `.items()`. T∆∞∆°ng t·ª± nh∆∞ `.keys()` v√† `.values()`, `.items()` tr·∫£ v·ªÅ m·ªôt view object g·ªçi l√† `dict_items` ch·ª©a m·ªôt 2-tuple cho m·ªói c·∫∑p `(key, value)`.

B·∫°n c√≥ th·ªÉ l·∫∑p qua t·ª´ng c·∫∑p v·ªõi v√≤ng l·∫∑p `for pair in <dict>.items():`, nh∆∞ng Python cho ph√©p b·∫°n *unpack* hai gi√° tr·ªã ƒë√≥ b√™n trong m·ªói tuple b·∫±ng c√°ch ch·ªâ ƒë·ªãnh hai bi·∫øn v√≤ng l·∫∑p. H√£y xem v√≤ng l·∫∑p d∆∞·ªõi ƒë√¢y:

```python
capitals = { 'Victoria': 'Melbourne',
             'New South Wales': 'Sydney' }

d_items = capitals.items()

print("Items in this dictionary:", d_items)
print("As a list:", list(d_items))

for key, value in d_items:
    print(key, value)
```

Slide ti·∫øp theo s·∫Ω ƒëi s√¢u h∆°n v·ªÅ unpacking.

## C√≥ √≠ch kh√¥ng?

B·∫°n s·∫Ω t√¨m th·∫•y nhi·ªÅu c√°ch s·ª≠ d·ª•ng kh√°c nhau cho `.items()` khi l√†m vi·ªác v·ªõi dictionary, v√¨ n√≥ cho ph√©p b·∫°n x·ª≠ l√Ω c√°c c·∫∑p `(key, value)` ri√™ng l·∫ª. V√≠ d·ª•, b·∫°n c√≥ th·ªÉ mu·ªën s·∫Øp x·∫øp c√°c keys theo *values* c·ªßa ch√∫ng. C√≥ nh·ªØng c√°ch thanh l·ªãch h∆°n, ch·∫≥ng h·∫°n nh∆∞ s·ª≠ d·ª•ng ƒë·ªëi s·ªë `key` trong `sorted()`, nh∆∞ng khi s·ª≠ d·ª•ng `.items()` b·∫°n ch·ªâ c·∫ßn ho√°n ƒë·ªïi th·ª© t·ª± c·ªßa keys v√† values, v√† s·∫Øp x·∫øp danh s√°ch tuple k·∫øt qu·∫£:

```python
my_dict = {"Melbourne": 5, "Sydney": 2, "Brisbane": 3}
swapped = []

for key, value in my_dict.items():
    swapped.append((value, key))

for value, key in sorted(swapped):
    print(key)
```

# ƒê·∫øm C√°c Ph·∫ßn T·ª≠ V·ªõi Dictionary

M·ªôt nhi·ªám v·ª• l·∫≠p tr√¨nh ph·ªï bi·∫øn l√† theo d√µi s·ªë l·∫ßn xu·∫•t hi·ªán c·ªßa c√°c m·ª•c kh√°c nhau trong m·ªôt t·∫≠p h·ª£p d·ªØ li·ªáu. D∆∞·ªõi ƒë√¢y l√† m·ªôt ch∆∞∆°ng tr√¨nh v√≠ d·ª• ƒë·∫øm s·ªë l·∫ßn xu·∫•t hi·ªán c·ªßa m·ªói ch·ªØ c√°i trong ƒëo·∫°n ƒë·∫ßu ti√™n c·ªßa Moby Dick:

```python
MOBY = """Call me Ishmael. Some years ago - never mind how long precisely - having little or no money in my purse, and nothing particular to interest me on shore, I thought I would sail about a little and see the watery part of the world. It is a way I have of driving off the spleen and regulating the circulation. Whenever I find myself growing grim about the mouth; whenever it is a damp, drizzly November in my soul; whenever I find myself involuntarily pausing before coffin warehouses, and bringing up the rear of every funeral I meet; and especially whenever my hypos get such an upper hand of me, that it requires a strong moral principle to prevent me from deliberately stepping into the street, and methodically knocking people's hats off - then, I account it high time to get to sea as soon as I can. This is my substitute for pistol and ball. With a philosophical flourish Cato throws himself upon his sword; I quietly take to the ship. There is nothing surprising in this. If they but knew it, almost all men in their degree, some time or other, cherish very nearly the same feelings towards the ocean with me."""
```

```python
tally = {}
```

```python
for char in MOBY:
    if char in tally:
        tally[char] += 1
    else:
        tally[char] = 1
```

```python
print(tally['C'])
print(tally['I'])
```
