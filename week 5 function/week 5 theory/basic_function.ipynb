# Hàm là gì?

Hàm giống như con dê cưng của bạn: nó có đầu vào (thức ăn), thực hiện một việc gì đó, và thường tạo ra đầu ra (...). Một cách khác để nhìn nhận là nó như một cách viết tắt tiện lợi cho một công thức. Bạn có thể đã thấy các hàm trong toán học: một hàm như f(x)=4x+5 có thể nhận một đầu vào (f(3) biểu thị giá trị 3 cho x), có một quá trình (giải biểu thức (4×3)+5 với đầu vào là 3), và một đầu ra (là kết quả của công thức đó: trong trường hợp này là 17). Cách sử dụng nó (với đầu vào trong dấu ngoặc) gần như giống hệt với các hàm trong Python, như chúng ta sẽ thấy.

Đến với một ví dụ cụ thể: giả sử chúng ta có công thức sau để tính số đồ uống tiêu chuẩn trong một ly:

```python
std_drinks = volume * percentage * 0.789
```

Ở đây `volume` là một biến (kích thước đồ uống của bạn tính bằng lít), và `percentage` là phần trăm cồn. Ví dụ, nếu bạn có một ly Long Island Iced Tea, có thể tích 0.375L và 13.5% cồn, bạn sẽ có:

```python
# Tính số đồ uống tiêu chuẩn
# chứa trong `volume` lít đồ uống có cồn
# với `percentage` % cồn
volume = 0.375
percentage = 13.5
drinks = volume * percentage * 0.789
print(drinks)
```

Rất nhiều đồ uống tiêu chuẩn. Nhưng nếu chúng ta muốn thực hiện phép tính này với các giá trị khác nhau của `volume` và `percentage` thì sao? Chắc chắn, có một cách để làm điều này mà không cần phải tái tạo lại đoạn code giống hệt nhau nhiều lần - giống như chúng ta có thể chạy các hàm có sẵn với các đầu vào khác nhau mà không cần phải lo lắng về các chi tiết phức tạp của cách chúng hoạt động? Câu trả lời là có, dưới dạng một hàm do người dùng định nghĩa:

```python
def std_drinks(volume, percentage):
    return volume * percentage * 0.789

print(std_drinks(0.375, 13.5))
print(std_drinks(0.586, 3.8))
print(std_drinks(1.5, 0))
```

Bài tập này tập trung vào cách định nghĩa và sử dụng các hàm do người dùng định nghĩa... một điều mà chúng ta sẽ làm rất nhiều trong phần còn lại của môn học này.

Hãy xem xét ví dụ của chúng ta chi tiết hơn:

```python
# Tính số lượng đồ uống chuẩn
# có trong `volume` lít đồ uống có cồn
# với `percentage` % cồn
def std_drinks(volume, percentage):
    return volume * percentage * 0.789
    
# Chúng ta sử dụng hàm để thực hiện tính toán
print(std_drinks(0.375, 13.5))
```

Tên hàm là `std_drinks`. `volume` và `percentage` được gọi là **tham số** (parameters). Chúng rất giống với các biến mà bạn đã thấy cho đến nay, và định nghĩa các đầu vào cho hàm. Chúng nhận các giá trị mà bạn cung cấp khi **gọi** (tức là sử dụng) hàm. Bản thân các giá trị được gọi là **đối số** (arguments). 

Hàm được gọi bằng cách đặt một cặp dấu ngoặc đơn `()` sau tên hàm, và đặt các đối số bên trong chúng.

Biểu thức sau `return` là những gì hàm sẽ xuất ra. Điều này có nghĩa là khi chúng ta viết biểu thức `std_drinks(0.375, 13.5)` trong mã của mình, nó sẽ **được tính toán** thành giá trị `return` của hàm, sẵn sàng để được gán cho một biến hoặc được `print` ra.

# Tại sao chúng ta nên sử dụng hàm (function)?

Điều tuyệt vời về các hàm là chúng ta có thể sử dụng chúng bao nhiêu lần tùy thích, với việc truyền vào các giá trị khác nhau cho các tham số. Hàm này rất đơn giản, nhưng nếu hàm phức tạp hơn, việc định nghĩa và gọi nó nhiều lần như một hàm sẽ tiết kiệm rất nhiều không gian! Khi lập trình, nếu bất cứ điều gì liên quan đến việc sử dụng quá nhiều copy và paste, chắc chắn phải có cách tốt hơn!

Ngoài hiệu quả của việc gọi hàm với các đầu vào khác nhau, nếu có lỗi trong hàm, chúng ta chỉ phải sửa một lần (trong định nghĩa hàm) thay vì ở mọi vị trí trùng lặp trong toàn bộ code.

```python
# Tính số lượng đồ uống chuẩn
# chứa trong `volume` lít đồ uống có cồn
# với `percentage` % cồn
def std_drinks(volume, percentage):
    return volume * percentage * 0.789

print(std_drinks(0.375, 13.5))
print(std_drinks(0.375, 5))
print(std_drinks(1.5, 0))
```

Lưu ý bạn cũng có thể đặt tên các đối số cùng với các giá trị trong lời gọi hàm và nếu làm như vậy, thứ tự của các tham số không còn quan trọng nữa: `std_drinks(percentage=4.5, volume=0.375)` giống với `std_drinks(0.375, 4.5)`. Hãy thử chỉnh sửa đoạn code trên để thử điều này.

Điều hữu ích khác về các hàm là chúng có **tên**. Thay vì phải nhớ một số dòng code làm gì, bạn có thể biến chúng thành một hàm để gắn nhãn chức năng của chúng. Điều này có nghĩa là code sẽ được tổ chức tốt hơn và có thể dễ dàng được tham chiếu theo tên.

# Cấu Trúc Của Một Định Nghĩa Hàm

Các thành phần cơ bản của một hàm bao gồm:
1. tên hàm;
2. các tham số;
3. thân hàm, nơi thực hiện tính toán thực tế liên quan đến hàm;
4. một cơ chế nào đó để trả về giá trị và thoát khỏi hàm.

Bạn có thể nhớ lại các hàm từ khi chúng ta giới thiệu **trừu tượng hóa** trong bài tập 0. Đây là nơi chúng ta sử dụng khối `def` chứa thân hàm để định nghĩa các hàm của mình, và khối `call` để chạy chúng, bao gồm các giá trị đối số sau đó nếu có.

Hãy cùng xem qua từng thành phần của một hàm khi áp dụng trong việc sử dụng mã Python.

Đầu tiên, để định nghĩa hàm của chúng ta, chúng ta sử dụng từ khóa `def`, theo sau là khoảng trắng, tên hàm, một cặp dấu ngoặc đơn, và dấu hai chấm (`:`). Ví dụ, nếu tên hàm của chúng ta là `my_first_function`, thì phần khung của định nghĩa hàm sẽ như sau:

```python
def my_first_function():
```

Tên hàm tuân theo chính xác các quy tắc giống như tên biến: chúng phải bắt đầu bằng một chữ cái hoặc dấu gạch dưới (`_`), và có thể được tạo thành từ bất kỳ sự kết hợp nào của các chữ cái, số và dấu gạch dưới... với một điều kiện nhỏ (như với các biến) là tên không thể là một từ dành riêng như `def` hoặc `if`.

Thứ hai, chúng ta chèn các tham số vào trong dấu ngoặc đơn trong định nghĩa hàm, được phân tách bằng dấu phẩy, ví dụ trong trường hợp ba đối số `num0`, `num1` và `num2`:

```python
def my_first_function(num0, num1, num2):
```

Điều thứ ba chúng ta làm trong định nghĩa hàm là liên kết đoạn code sẽ được thực thi mỗi khi hàm được gọi, dưới dạng một **khối code**, được thụt lề bằng khoảng trắng từ header của hàm, giống như cách chúng ta liên kết một khối code với câu lệnh `if`, `elif` hoặc `else`. Lưu ý rằng tất nhiên có thể có một câu lệnh `if` "lồng nhau" trong một hàm, và chúng ta chỉ cần thụt lề khối code liên quan xa hơn nữa, so với thân của hàm (có nghĩa là càng lồng code sâu trong Python, chúng ta càng biến mất về phía bên phải của trang!). Giống như với câu lệnh `if`, bốn dấu cách được khuyến khích cho việc thụt lề của câu lệnh `def`. Giả sử chúng ta định nghĩa một hàm tìm số lớn nhất trong ba số:

```python
def my_first_function(num0, num1, num2):
    maxnum = num0
    if num1 > maxnum:
        maxnum = num1
    if num2 > maxnum:
        maxnum = num2
```

Đây là một định nghĩa hàm hợp lệ, nhưng chúng ta đang thiếu thành phần quan trọng thứ tư: một cách để định nghĩa kết quả của hàm là gì. Điều này được thực hiện bằng câu lệnh `return`, câu lệnh này nhận một giá trị duy nhất, ví dụ:

```python
def my_first_function(num0, num1, num2):
    maxnum = num0
    if num1 > maxnum:
        maxnum = num1
    if num2 > maxnum:
        maxnum = num2
    return maxnum
```

Khi bạn chạy đoạn code này, điều gì xảy ra? Không có gì? Thực ra, Python đang thực thi định nghĩa hàm để hàm có thể được gọi, nhưng trong trường hợp không có lời gọi hàm thực tế, định nghĩa không bao giờ thực sự được sử dụng. Chúng ta sẽ thấy các ví dụ về cả định nghĩa hàm và lời gọi hàm trong slide tiếp theo.

**`return` là một câu lệnh hay một hàm?** `return` là một **câu lệnh** theo nghĩa là các đối số liên quan được cung cấp ở phía bên phải của nó, được phân tách bằng khoảng trắng (giống như các đối số cho `if` được cung cấp trực tiếp mà không có dấu ngoặc đơn nào). Điều này khác với các hàm như `len()`, trong đó các đối số được cung cấp trong dấu ngoặc đơn. Câu lệnh khác với hàm ở chỗ chúng thực hiện một hành động trực tiếp, trong khi hàm thực hiện một tính toán và (có thể) trả về một giá trị. Với nguy cơ sa lầy vào ngữ nghĩa, `return` là một câu lệnh vì bạn đang ra lệnh cho Python thực hiện điều gì đó trực tiếp (cụ thể là thoát khỏi hàm và trả về một giá trị).

# Gọi Hàm Do Người Dùng Định Nghĩa

Sau khi đã định nghĩa một hàm, chúng ta có thể tiến hành gọi nó, theo cách hoàn toàn giống như các hàm có sẵn, ví dụ:

```python
def my_first_function(num0, num1, num2):
    maxnum = num0
    if num1 > maxnum:
        maxnum = num1
    if num2 > maxnum:
        maxnum = num2
    return(maxnum)

print(my_first_function(1, 2.0, 3))
print(my_first_function(4, -1, 0))
print(my_first_function(3, 5, 1))
```

Có một vài điều cần lưu ý tại thời điểm này. Thứ nhất, một hàm phải được định nghĩa trước khi chúng ta có thể gọi nó; cách dễ nhất để đảm bảo điều này là để định nghĩa hàm đứng trước bất kỳ lời gọi hàm nào trong code (ví dụ, hãy thử chạy đoạn code trên với các lời gọi hàm ở những dòng trên định nghĩa hàm, và xem điều gì sẽ xảy ra).

Thứ hai, mỗi lần chúng ta gọi hàm, nó được tính toán với các đối số được cung cấp. Trong ví dụ trên của chúng ta, ví dụ, lần đầu tiên chúng ta gọi hàm, code được thực thi với `1` được gán cho `num0`; lần thứ hai, nó được thực thi với `4` được gán cho `num0`. Hai lời gọi khác nhau này không có kết nối với nhau: Python sẽ không nhớ các giá trị đối số của những lần gọi hàm trước đó trong những lần gọi tiếp theo.

# Ghi chú Quan trọng về Return

Khi câu lệnh `return` được thực thi trong một hàm, hàm sẽ dừng ngay lập tức và trả về giá trị đã cho. Điều này có nghĩa là bất kỳ đoạn code nào trong hàm sau câu lệnh `return` sẽ không được thực thi. Hãy thử dự đoán xem đoạn code sau sẽ xuất ra gì:

```python
def a_function():
    print("Returning a number:")
    return 1
    print("Returning another number:")
    return 3

print(a_function())
```

**Kết quả:**
```
Returning a number:
1
```

Từ ví dụ này, bạn có thể nghĩ rằng mỗi hàm chỉ có thể có một câu lệnh `return`. Không đúng: việc có nhiều câu lệnh `return` có thể rất hữu ích, nếu chúng được thực thi trong các trường hợp khác nhau. Ví dụ, chúng có thể nằm trong các nhánh khác nhau của câu lệnh `if..elif..else`.

Điểm quan trọng khác về `return` là không nhất thiết phải có giá trị trả về! Trong tất cả các ví dụ trước đây, chúng ta đã sử dụng `return` để truyền một giá trị đầu ra trở lại code gọi hàm. Tuy nhiên, nếu chúng ta viết một hàm **thực hiện một việc gì đó** nhưng không cần truyền bất kỳ đầu ra nào, chúng ta có thể sử dụng `return` mà không có giá trị đính kèm để đơn giản dừng hàm tại thời điểm đó, đưa việc thực thi trở lại code gọi hàm. Có tồn tại hàm hữu ích mà không trả về đầu ra không? Tất nhiên rồi, ví dụ như `print()`: giá trị trả về? Không. Hữu ích? Rất nhiều.

```python
def print_twice(output):
    print(output)
    print(output)
    return

print_twice("Hello there")
```

Đối với một hàm như thế này, bạn có thể bỏ qua `return` hoàn toàn và hàm sẽ dừng khi nó đạt đến cuối. Tuy nhiên đôi khi, như trong ví dụ trước, việc dừng thực thi hàm tại một điểm nhất định là rất quan trọng: một câu lệnh `return` rõ ràng rất hữu ích trong trường hợp này.

# Tổng kết về Hàm (Function Summary)

Trong các slide vừa qua, chúng ta đã chính thức giới thiệu về hàm và thực hành cách định nghĩa hàm của riêng mình.

* Một hàm nhận đầu vào, xử lý một số tác vụ và trả về kết quả đầu ra.
* Hàm cho phép một khối mã được chạy mà không cần phải viết lại mỗi lần. Điều này góp phần tạo ra mã sạch hơn và tổ chức chức năng một cách logic hơn.
* Một hàm được sử dụng (**gọi**) bằng cách viết tên của nó theo sau là một cặp dấu ngoặc đơn `()`. Các **đối số** đầu vào (nếu có) được viết bên trong dấu ngoặc, cách nhau bằng dấu phẩy. Giá trị trả về của hàm (nếu có) là thứ mà lời gọi hàm sẽ **được đánh giá thành** và sau đó có thể được gán cho một biến hoặc được `print` hoặc bất kỳ thứ gì khác.

Đây là cấu trúc của một định nghĩa hàm:

```python
def <tên hàm>(<tham số 1>, <tham số 2>):
    <khối mã>
```

Các thành phần bao gồm:

## 1. Tên hàm
* Được đặt sau từ khóa `def`.
* Quy tắc cho tên hàm hợp lệ giống như quy tắc cho tên biến hợp lệ.

## 2. Các tham số
* Được đặt bên trong dấu ngoặc đơn, cách nhau bằng dấu phẩy. Có thể có bao nhiêu tùy ý, hoặc không có.

## 3. Thân hàm
* Được viết dưới dạng một khối mã sau dòng `def`, được thụt lề bằng bốn dấu cách (được khuyến nghị).

## 4. Cơ chế trả về kết quả từ hàm
* Được thực hiện bởi câu lệnh `return`, trong đó giá trị sau từ khóa `return` trở thành kết quả mà hàm được đánh giá thành khi nó được gọi.
* Khi gặp câu lệnh `return`, hàm sẽ kết thúc. Nếu không có câu lệnh nào được gặp, hàm sẽ thực thi đến cuối.
* Nếu có nhiều câu lệnh `return`, chỉ câu lệnh đầu tiên được thực thi mới có hiệu lực: tại thời điểm này hàm sẽ kết thúc.
