# Hàm `between_len`

Viết một hàm `between_len` nhận ba tham số:
1. một danh sách `words`;
2. một số nguyên `minlen`; và
3. một số nguyên `maxlen`.

Hàm này sẽ trả về `True` nếu độ dài của `words` (tức là độ dài của *danh sách*, không phải độ dài của các từ trong danh sách) ít nhất là `minlen` và không quá `maxlen`, ngược lại trả về `False`.

Đối với những ai thích toán học, nếu lwords là độ dài của danh sách, chúng ta đang đánh giá tính đúng sai của biểu thức minlen≤lwords≤maxlen.

Dưới đây là các ví dụ gọi hàm (hiển thị kết quả với lệnh `print()` trong mỗi trường hợp), để minh họa chức năng của nó:

```python
>>> between_len(["python", "is", "boring"], 4, 7)
```

```
False
```

```python
>>> between_len(["python", "is", "amazing"], -1, 20)
```

```
True
```

```python
>>> between_len(["python", "is", "amazing"], minlen=3, maxlen=4)
```

```
True
```

## **Trả về giá trị Boolean**

Vì sử dụng toán tử quan hệ tạo ra giá trị Boolean, một cách thanh lịch để trả về Boolean là trả về điều kiện trực tiếp. Ví dụ, nếu chúng ta muốn kiểm tra xem một số `n` có lớn hơn `5` hay không, đây là ví dụ trả về giá trị của phép kiểm tra đó trực tiếp (*très élégante!*).

```python
def isGreater(n):
    return n > 5
```

Đoạn code sau có chức năng tương đương (tức là làm chính xác cùng một việc như đoạn đầu tiên), nhưng với bước thừa không cần thiết là kiểm tra kết quả của phép kiểm tra, và trả về Boolean tương ứng một cách rõ ràng:

```python
def isGreater(n):
    if n > 5:
        return True
    else:
        return False
```

# Đếm Từ (Word Counter)

Khám phá các phương thức cho chuỗi trong terminal bằng cách sử dụng `dir("hello world")`. Với những kiến thức này, hãy viết một hàm `word_count(input_str)` nhận một chuỗi làm đầu vào và trả về số từ trong chuỗi đó. Hàm của bạn nên hoạt động như sau:

>>> word_count("There are 44 cows on a hill")

```

```

```

```
